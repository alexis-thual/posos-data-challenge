```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

from functools import partial
from scipy import sparse
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
colors = ["#2965CC", "#29A634", "#D99E0B", "#D13913", "#8F398F", "#00B3A4", "#DB2C6F", "#9BBF30", "#96622D", "#7157D9"]
```

# Chargement des donn√©es

```python
dataFolder = './challenge_data'
xPath = os.path.join(dataFolder, 'input_train.csv')
yPath = os.path.join(dataFolder, 'challenge_output_data_training_file_predict_the_expected_answer.csv')

# Loading, parsing and spliting training and testing data
x = pd.read_csv(xPath, delimiter=';', usecols=[0, 1]).values
y = pd.read_csv(yPath, delimiter=';', usecols=[0, 1]).values
```

# Distribution du vocabulaire

```python
def parseWord(wordCount, item):
    return [item[0], wordCount[0, item[1]]]
```

```python
countVectorizer = CountVectorizer(input='content')
countVector = countVectorizer.fit_transform(x[:,1])
print(countVector.shape)
vocabulary = countVectorizer.vocabulary_

wordCount = np.sum(countVector, axis=0)
print(wordCount.shape)
vocabulary = list(map(partial(parseWord, wordCount), vocabulary.items()))
```

```python
sortedVocabulary = sorted(vocabulary, key=lambda x: x[1], reverse=True)
sortedVocabulary[:5]
```

```python
numberWords = 50

labels = [voc[0] for voc in sortedVocabulary[:numberWords]]
values = [voc[1] for voc in sortedVocabulary[:numberWords]]
yPosition = np.arange(len(labels))

plt.figure(figsize=(15,8))
plt.bar(yPosition, values, color=colors[0])
plt.xticks(yPosition, labels, rotation=90)
plt.ylabel('Unique count')
plt.title('Word')
 
plt.show()
```

# Distribution des categories

```python
numberCategories = len(set(y[:,1]))
countCategories = [0] * numberCategories

for point in y:
    countCategories[point[1]] += 1

indexedCountCategories = list(zip(list(range(numberCategories)), countCategories))
```

```python
yPosition = np.arange(numberCategories)

plt.figure(figsize=(15,8))
plt.bar(yPosition, countCategories, color=colors[0])

plt.ylabel('Unique count')
plt.title('Word')
 
plt.show()
```

```python
selectedCategories = list(filter(lambda x: x[1] > 100, indexedCountCategories))

labels = [category[0] for category in selectedCategories]
values = [category[1] for category in selectedCategories]
yPosition = np.arange(len(selectedCategories))

plt.figure(figsize=(15,8))
plt.bar(yPosition, values, color=colors[0])
plt.xticks(yPosition, labels, rotation=90)
plt.ylabel('Unique count')
plt.title('Word')
 
plt.show()

print('Number of categories with more than 100 points: {}'.format(str(len(selectedCategories))))
```
